```python
# includes
# %matplotlib inline
% autosave 15

import matplotlib.pyplot as plt
#import seaborn as sns
import numpy as np
import pandas as pd
from Bio import SeqIO
```

```python
gene_names = []
with open('../data/clean/names') as input_file:
    for line in input_file:
        gene_names.append(line[:-1].split(','))
print(gene_names)
```

```python
def fasta_plot(filename, destdir):
    raw_data = SeqIO.parse(open('../data/raw/' + filename + '.s.fasta'), 'fasta')
    genes_processed = 0
    for fasta in raw_data:
        if genes_processed == 10: 
            break
        name = fasta.id,
        feature_vec = pd.DataFrame([float(val) for val in str(fasta.seq).split(',')])
        plt.figure(figsize=(50,30))
        plt.ylim(ymin=0, ymax=1.2)
        plt.xlabel(str(name), fontsize=60)
        plt.ylabel(filename, fontsize=60)
        plt.plot(feature_vec, maker='o', linewidth=3)
        plt.savefig('../images/' + destdir + '/' + str(genes_processed) +'.png')
        plt.close()
        genes_processed += 1
```

```python
fasta_plot('CA', '1_CA')
fasta_plot('CG', '2_CG')
fasta_plot('COUNT_SNP', '3_SNP_CNT')
fasta_plot('COV', '4_COV')
fasta_plot('MAF_SNP', '5_MAF')
#fasta_plot('CA', '6_METHYL')
fasta_plot('TATA', '7_TATA')
fasta_plot('TFBS.inform', '8_TFBS_INF')
fasta_plot('TFBS', '9_TFBS')
```
